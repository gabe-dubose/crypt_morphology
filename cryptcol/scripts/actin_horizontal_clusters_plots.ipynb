```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data
data = pd.read_csv('../processed_data/horizontal_actin_clusters.csv')
```

```python
data
```

```python
sns.set_style('whitegrid')
fig, [ax1,ax2,ax3,ax4] = plt.subplots(1, 4, figsize=(16, 4), sharey=True)

value_cols = ['idx1', 'idx2', 'idx3', 'idx4', 'idx5', 'idx6', 'idx7', 'idx8', 'idx9', 'idx10']

#cluster1
cluster_1 = data.loc[data['cutree.actin.clusters..h...1.95.'] == 1]
clust1_samples = list(cluster_1['sample_id'])
clust1_data = cluster_1[value_cols]
for index, row in clust1_data.iterrows():
    values = list(row)
    sns.lineplot(values, ax=ax1, color='tab:gray', alpha=0.5)

cluster_2 = data.loc[data['cutree.actin.clusters..h...1.95.'] == 2]
clust2_samples = list(cluster_2['sample_id'])
clust2_data = cluster_2[value_cols]
for index, row in clust2_data.iterrows():
    values = list(row)
    sns.lineplot(values, ax=ax2, color='tab:gray', alpha=0.8)

cluster_3 = data.loc[data['cutree.actin.clusters..h...1.95.'] == 3]
clust3_samples = list(cluster_3['sample_id'])
clust3_data = cluster_3[value_cols]
for index, row in clust3_data.iterrows():
    values = list(row)
    sns.lineplot(values, ax=ax3, color='tab:gray', alpha=0.8)


cluster_4 = data.loc[data['cutree.actin.clusters..h...1.95.'] == 4]
clust4_samples = list(cluster_4['sample_id'])
clust4_data = cluster_4[value_cols]
for index, row in clust4_data.iterrows():
    values = list(row)
    sns.lineplot(values, ax=ax4, color='tab:gray', alpha=0.8)


#add mean value lines
sns.lineplot(list(clust1_data.mean(axis=0)), ax=ax1, color='black', linewidth=3)
sns.lineplot(list(clust2_data.mean(axis=0)), ax=ax2, color='black', linewidth=3)
sns.lineplot(list(clust3_data.mean(axis=0)), ax=ax3, color='black', linewidth=3)
sns.lineplot(list(clust4_data.mean(axis=0)), ax=ax4, color='black', linewidth=3)

plt.savefig('../figures/actin_clusters.pdf')
```
