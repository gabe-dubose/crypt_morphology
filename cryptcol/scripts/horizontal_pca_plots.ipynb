```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read pca data
total_data = pd.read_csv('../processed_data/horizontal_total_pca.csv')
rfp_data = pd.read_csv('../processed_data/horizontal_rfp_pca.csv')
actin_data = pd.read_csv('../processed_data/horizontal_actin_pca.csv')

#load cluster data
rfp_clusters = pd.read_csv('../processed_data/horizontal_rfp_clusters.csv')
actin_clusters = pd.read_csv('../processed_data/horizontal_actin_clusters.csv')
```

```python
#I'll start with total data
stain = []
position = []
for sample_id in list(total_data['data.sample_id']):
    if 'actin' in sample_id:
        stain.append('Actin')
    else:
        stain.append('RFP')
        
    if 'medial' in sample_id:
        position.append('Medial')
    elif 'posterior' in sample_id:
        position.append('Posterior')
    elif 'anterior' in sample_id:
        position.append('Anterior')
        
total_data['Position'] = position
total_data['Type'] = stain
```

### Total Data

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(1, 1, figsize=(6, 5))

sns.scatterplot(data=total_data, x='PC1', y='PC2', hue='Position', style='Type', ax=ax, s=75)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set_xlabel('PC1: 71.24%')
ax.set_ylabel('PC2: 16.28%')
ax.set_title('Total')

plt.tight_layout()
plt.savefig('../figures/total_pca.pdf')
```

### RFP Data

```python
#I'll start with total data
clusters = []
position = []
for sample_id in list(rfp_data['rfp.data.sample_id']):        
    if 'medial' in sample_id:
        position.append('Medial')
    elif 'posterior' in sample_id:
        position.append('Posterior')
    elif 'anterior' in sample_id:
        position.append('Anterior')
    
    cluster = str(int(rfp_clusters.loc[rfp_clusters['sample_id'] == sample_id]['cutree.rfp.clusters..h...1.85.']))
    clusters.append(cluster)
        
rfp_data['Position'] = position
rfp_data['Cluster'] = clusters
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(1, 1, figsize=(6, 5))

sns.scatterplot(data=rfp_data, x='PC1', y='PC2', hue='Position', ax=ax, s=75)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set_xlabel('PC1: 62.85%')
ax.set_ylabel('PC2: 23.93%')
ax.set_title('RFP')

plt.tight_layout()
plt.savefig('../figures/rfp_pca_by_position.pdf')
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(1, 1, figsize=(6, 5))

sns.scatterplot(data=rfp_data, x='PC1', y='PC2', hue='Cluster', ax=ax, s=75)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set_xlabel('PC1: 62.85%')
ax.set_ylabel('PC2: 23.93%')
ax.set_title('RFP')

plt.tight_layout()
plt.savefig('../figures/rfp_pca_by_cluster.pdf')
```

### Actin Data

```python
#I'll start with total data
clusters = []
position = []
for sample_id in list(actin_data['actin.data.sample_id']):        
    if 'medial' in sample_id:
        position.append('Medial')
    elif 'posterior' in sample_id:
        position.append('Posterior')
    elif 'anterior' in sample_id:
        position.append('Anterior')
    
    cluster = str(int(actin_clusters.loc[actin_clusters['sample_id'] == sample_id]['cutree.actin.clusters..h...1.95.']))
    clusters.append(cluster)
        
actin_data['Position'] = position
actin_data['Cluster'] = clusters
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(1, 1, figsize=(6, 5))

sns.scatterplot(data=actin_data, x='PC1', y='PC2', hue='Position', ax=ax, s=75)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set_xlabel('PC1: 45.61%')
ax.set_ylabel('PC2: 27.43%')
ax.set_title('Actin')

plt.tight_layout()
plt.savefig('../figures/actin_pca_by_position.pdf')
```
